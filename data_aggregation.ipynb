```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

fbref_data = pd.read_csv('./data/new_matches.csv')
fbref_data.columns = fbref_data.columns.str.lower()
fbref_data['date'] = pd.to_datetime(fbref_data['date'])
fbref_data
```

```python
df_17 = pd.read_csv('./data/17_18.csv')
df_18 = pd.read_csv('./data/18_19.csv')
df_19 = pd.read_csv('./data/19_20.csv')
df_20 = pd.read_csv('./data/20_21.csv')
df_21 = pd.read_csv('./data/21_22.csv')
df_22 = pd.read_csv('./data/22_23.csv')
df_23 = pd.read_csv('./data/23_24.csv')
df_24 = pd.read_csv('./data/24_25.csv')
df_17 = df_17[['Date', 'HomeTeam', 'AwayTeam', 'B365H', 'B365D', 'B365A', 'BbAv>2.5', 'BbAv<2.5']]
df_18 = df_18[['Date', 'HomeTeam', 'AwayTeam', 'B365H', 'B365D', 'B365A', 'BbAv>2.5', 'BbAv<2.5']]
df_19 = df_19[['Date', 'HomeTeam', 'AwayTeam', 'B365H', 'B365D', 'B365A', 'B365>2.5', 'B365<2.5']]
df_20 = df_20[['Date', 'HomeTeam', 'AwayTeam', 'B365H', 'B365D', 'B365A', 'B365>2.5', 'B365<2.5']]
df_21 = df_21[['Date', 'HomeTeam', 'AwayTeam', 'B365H', 'B365D', 'B365A', 'B365>2.5', 'B365<2.5']]
df_22 = df_22[['Date', 'HomeTeam', 'AwayTeam', 'B365H', 'B365D', 'B365A', 'B365>2.5', 'B365<2.5']]
df_23 = df_23[['Date', 'HomeTeam', 'AwayTeam', 'B365H', 'B365D', 'B365A', 'B365>2.5', 'B365<2.5']]
df_24 = df_24[['Date', 'HomeTeam', 'AwayTeam', 'B365H', 'B365D', 'B365A', 'B365>2.5', 'B365<2.5']]
print(len(df_17), len(df_18), len(df_19), len(df_20), len(df_21), len(df_22), len(df_23), len(df_24))
```

```python
df_17.rename(columns={'BbAv>2.5' : 'over2.5', 'BbAv<2.5' : 'under2.5'}, inplace=True)
df_18.rename(columns={'BbAv>2.5' : 'over2.5', 'BbAv<2.5' : 'under2.5'}, inplace=True)
df_19.rename(columns={'B365>2.5' : 'over2.5', 'B365<2.5' : 'under2.5'}, inplace=True)
df_20.rename(columns={'B365>2.5' : 'over2.5', 'B365<2.5' : 'under2.5'}, inplace=True)
df_21.rename(columns={'B365>2.5' : 'over2.5', 'B365<2.5' : 'under2.5'}, inplace=True)
df_22.rename(columns={'B365>2.5' : 'over2.5', 'B365<2.5' : 'under2.5'}, inplace=True)
df_23.rename(columns={'B365>2.5' : 'over2.5', 'B365<2.5' : 'under2.5'}, inplace=True)
df_24.rename(columns={'B365>2.5' : 'over2.5', 'B365<2.5' : 'under2.5'}, inplace=True)
fb_data = pd.concat([df_17, df_18, df_19, df_20, df_21, df_22, df_23, df_24])
fb_data.columns = fb_data.columns.str.lower()
fb_data
```

```python
fb_data['date'] = pd.to_datetime(fb_data['date'], format='%d/%m/%Y')
fb_data
```

```python
team_name_map = {
    'Man City' : 'Manchester City',
    'Nott\'m Forest' : 'Nottingham Forest',
    'Tottenham' : 'Tottenham Hotspur',
    'Brighton' : 'Brighton and Hove Albion',
    'Newcastle' : 'Newcastle United',
    'Man United' : 'Manchester United',
    'West Ham' : 'West Ham United',
    'Leicester' : 'Leicester City',
    'Ipswich' : 'Ipswich Town',
    'Wolves' : 'Wolverhampton Wanderers',
    'Luton' : 'Luton Town',
    'Leeds' : 'Leeds United',
    'Norwich' : 'Norwich City',
    'West Brom' : 'West Bromwich Albion',
    'Cardiff' : 'Cardiff City',
    'Huddersfield' : 'Huddersfield Town',
    'Swansea' : 'Swansea City',
    'Stoke' : 'Stoke City'
}
```

```python
fb_data['hometeam'] = fb_data['hometeam'].replace(team_name_map)
fb_data['awayteam'] = fb_data['awayteam'].replace(team_name_map)
fb_data
```

```python
fb_data = pd.concat([fb_data]*2, ignore_index=True)
fb_data
```

```python
merge_data = pd.merge(
    fb_data, fbref_data, left_on=['date', 'hometeam'], right_on=['date', 'team'], suffixes=('', '_home')
)
merge_data
```
