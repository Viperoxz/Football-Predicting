```python
df['round'] = df['round'].str.replace('Matchweek ', '')
df['date'] = pd.to_datetime(df['date'])
df.drop(columns=['comp'], inplace=True)
df.head()
```

```python
df['venue'] = df['venue'].astype('category').cat.codes
df['opponent'] = df['opponent'].astype('category').cat.codes
```

```python
df['time'] = df['time'].str.replace(':.+', '', regex=True).astype(int)
df['day'] = df['date'].dt.dayofweek
df.head()
```

```python
df['target'] = df['result'].apply(lambda x: 1 if x == 'W' else 0)
df.head()
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators=50, min_samples_split=10, random_state=42)
```

```python
predictors = ['venue', 'opponent', 'time', 'day']
```

```python
rf.fit(train[predictors], train['target'])
```

```python
preds = rf.predict(test[predictors])
```

```python
def rolling_avg(group, cols, new_cols):
    group = group.sort_values('date')
    rolling_stats = group[cols].rolling(3, closed='left').mean()
    group[new_cols] = rolling_stats
    group = group.dropna(subset=new_cols)
    return group
```

```python
def make_predictions(data, predictors):
    train = data[data['date'] < '2022-01-01']
    test = data[data['date'] >= '2022-01-01']
    rf.fit(train[predictors], train['target'])
    preds = rf.predict(test[predictors])
    combined = pd.DataFrame(dict(actual=test['target'], predicted=preds), index=test.index)
    precision = precision_score(test['target'], preds)
    return combined, precision
```

```python
# Số kỳ thủ
num_players = 8

# Danh sách trận đấu chưa diễn ra
matches = [
    (1, 2), (1, 5), (1, 6), (1, 7), (1, 8),
    (2, 3), (2, 5), (2, 6), (2, 7), (2, 8),
    (3, 5), (3, 6), (3, 7), (3, 8),
    (4, 5), (4, 6),
    (5, 7),
    (6, 7), (6, 8),
    (7, 8)
]

# Đồ thị để lưu các cạnh giữa các trận có chung kỳ thủ
graph = {match: [] for match in matches}

# Tạo cạnh giữa các trận có chung kỳ thủ
for i in range(len(matches)):
    for j in range(i + 1, len(matches)):
        match1 = matches[i]
        match2 = matches[j]
        if match1[0] in match2 or match1[1] in match2:
            graph[match1].append(match2)
            graph[match2].append(match1)

# Dùng thuật toán tham lam để tô màu các đỉnh (trận đấu) của đồ thị
colors = {}
for match in matches:
    # Tìm màu khả dụng
    used_colors = {colors[neighbor] for neighbor in graph[match] if neighbor in colors}
    # Chọn màu nhỏ nhất không bị trùng
    color = 1
    while color in used_colors:
        color += 1
    colors[match] = color

# Kết quả: Số buổi thi đấu là số màu lớn nhất
num_sessions = max(colors.values())
print("Số buổi thi đấu tối thiểu:", num_sessions)

# Hiển thị lịch thi đấu
schedule = [[] for _ in range(num_sessions)]
for match, color in colors.items():
    schedule[color - 1].append(match)

for i, session in enumerate(schedule, 1):
    print(f"Buổi {i}: {session}")
```
