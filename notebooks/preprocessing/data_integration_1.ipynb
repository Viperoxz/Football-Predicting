```python
def fill_na_opponent(df, type):
    opp = 'awayteam' if type == 'hometeam' else 'hometeam'
    for index, row in df[df['opponent'].isnull()].iterrows():
        team = row['team']
        match_date = row['date']
        possible_opponents = fb_data[(fb_data['date'] == match_date) & 
                                    (fb_data[type] == team)]
        if not possible_opponents.empty:
            opponent = possible_opponents.iloc[0][opp]
            df.at[index, 'opponent'] = opponent
    return df
```

```python
fbref_data_home =  fill_na_opponent(fbref_data_home, 'hometeam')
fbref_data_away =  fill_na_opponent(fbref_data_away, 'awayteam')
```
