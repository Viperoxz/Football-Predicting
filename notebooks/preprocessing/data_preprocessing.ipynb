```python
import pandas as pd
```

```python
df = pd.read_csv('../../data/raw/fbdata_fbref.csv')
df.head()
```

```python
df.drop(columns=['hometeam', 'awayteam', 'comp', 'match report', 'notes'], inplace=True)
df['round'] = df['round'].str.extract('(\d+)').astype(int)
df['date'] = pd.to_datetime(df['date'])
df['day'] = df['date'].dt.dayofweek
df['time'] = df['time'].str.replace(':.+', '', regex=True).astype(int)
df['venue'] = df['venue'].astype('category').cat.codes
df
```

```python
# move column opponent to next to team
cols = ['team', 'opponent', 'venue'] + [col for col in df if col not in ['team', 'opponent', 'venue']]
df = df[cols]
df.head()
```

```python
df['team'].value_counts()
```

```python
result_map = {'W': 3, 'D': 1, 'L': 0}
df['result'] = df['result'].map(result_map)
```

```python

# get the points of a team after a certain round in a season
def get_points(team, season, round):
    season_games = df[(df['team'] == team) & (df['season'] == season) & (df['round'] <= round)]
    if season_games.empty:
        return 0 
    points = season_games['result'].sum()
    return points

get_points('Manchester United', '2024-2025', 10)
```

```python
df['points'] = df.apply(lambda x: get_points(x['team'], x['season'], x['round']), axis=1)
df['points_before_match'] = df.apply(lambda x: get_points(x['team'], x['season'], x['round']) - x['result'], axis=1)
df.head(10)
```

```python
def fix_formation(formation):
    parts = formation.split('-')[:-1]
    last = 10 - sum(int(num) for num in parts)
    formation = '-'.join(parts)
    if last:
        formation += f'-{last}'
    return formation

df['formation'] = df['formation'].apply(fix_formation)
df['formation'].value_counts()
```

```python
# arrange team on the order of points over all seasons 
team_order = df.groupby('team')['points'].sum().sort_values(ascending=False).index
team_order
```

```python
# base on team_order to encode team
team_map = {team: i for i, team in enumerate(team_order)}
team_map
```

```python
df['team'] = df['team'].map(team_map)
df['opponent'] = df['opponent'].map(team_map)
df
```

```python
# Because in df, a match is counted twice => just get df[df['venue] == 0 or 1] for calculate about referee, formation,...
sub_df = df[df['venue'] == 1]
```

```python
# get n/N * avg(gf - ga) for each formation
def pre_formation(formation):
    N = len(sub_df)
    games = sub_df[sub_df['formation'] == formation]
    n = len(games)
    return n/N * (games['gf'].mean() - games['ga'].mean())

pre_formation('4-3-3')
```

```python
formations = df['formation'].unique()
formation_rank = {formation: pre_formation(formation) for formation in formations}
formation_rank = {k: v for k, v in sorted(formation_rank.items(), key=lambda item: item[1], reverse=True)}
formation_rank
```

```python
formation_order = list(formation_rank.keys())
formation_map = {formation: i for i, formation in enumerate(formation_order)}
df['formation'] = df['formation'].map(formation_map)
df['opp formation'] = df['opp formation'].map(formation_map)
```

```python
# get n/N * avg(gf - ga) of a certain referee
def pre_referee(referee):
    N = len(sub_df)
    games = sub_df[(sub_df['referee'] == referee)]
    n = len(games)
    return n/N * (games['gf'].mean() - games['ga'].mean())

pre_referee('Anthony Taylor')
```

```python
referees = sub_df['referee'].unique()
referee_rank = {referee: pre_referee(referee) for referee in referees}
referee_rank = {k: v for k, v in sorted(referee_rank.items(), key=lambda item: item[1], reverse=True)}
referee_rank
```

```python
referee_map = {referee: i for i, referee in enumerate(referee_rank.keys())}
df['referee'] = df['referee'].map(referee_map)
```
